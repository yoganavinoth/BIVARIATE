```python
#variance inflation factor(VIF)
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    #claculating VIF
    vif=pd.DataFrame()
    vif["variable"]=X.columns
    vif["VIF"]=[variance_inflation_factor(X.values,i)for i in range(X.shape[1])]
    return (vif)
```

```python
from scipy.stats import ttest_ind
#drop null values
dataset=dataset.dropna()
male=dataset[dataset["gender"]=="M"]["salary"]
female=dataset[dataset["gender"]=="F"]["salary"]
#print(male)
ttest_ind(male,female)
```

```python
import scipy.stats as stats
stats.f_oneway(dataset["ssc_p"],dataset["hsc_p"],dataset["degree_p"])
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
# Two-way ANOVA
ano=ols("mba_p~salary+status",data=dataset).fit()
two=sm.stats.anova_lm(ano,type=2)
```
